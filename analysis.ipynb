```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a professional style for the plots
sns.set_theme(style="whitegrid")

# --- Step 1: Load the Results Data ---
try:
    df = pd.read_csv("results/results_log.csv")
    print("Results file loaded successfully.")
    print(df.head())
except FileNotFoundError:
    print("Error: results/results_log.csv not found. Make sure you have run the experiments.")

# --- Step 2: Plot Loss (MSE) Comparison ---
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='round', y='loss', hue='model_name', marker='o')
plt.title('Model Comparison: Loss (MSE) per Round', fontsize=16)
plt.xlabel('Federated Round', fontsize=12)
plt.ylabel('Loss (Mean Squared Error)', fontsize=12)
plt.legend(title='Model')
plt.xticks(df['round'].unique())
plt.grid(True)
plt.savefig("results/comparison_loss.png") # Save the plot
plt.show()


# --- Step 3: Plot MAE Comparison ---
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='round', y='mae', hue='model_name', marker='o')
plt.title('Model Comparison: MAE per Round', fontsize=16)
plt.xlabel('Federated Round', fontsize=12)
plt.ylabel('Mean Absolute Error', fontsize=12)
plt.legend(title='Model')
plt.xticks(df['round'].unique())
plt.grid(True)
plt.savefig("results/comparison_mae.png") # Save the plot
plt.show()


# --- Step 4: Plot R Score Comparison ---
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='round', y='r2', hue='model_name', marker='o')
plt.title('Model Comparison: R Score per Round', fontsize=16)
plt.xlabel('Federated Round', fontsize=12)
plt.ylabel('R Score', fontsize=12)
plt.legend(title='Model')
plt.xticks(df['round'].unique())
plt.grid(True)
plt.savefig("results/comparison_r2.png") # Save the plot
plt.show()
```
